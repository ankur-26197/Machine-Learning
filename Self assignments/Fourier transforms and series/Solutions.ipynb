## Question 1

```python
import numpy as np
import matplotlib.pyplot as plt

N=100
fxy=[1,10,20]
multiplier=10

def Q(x):
    return np.ceil(((x+1)/2)*255)

def dft_computation(image):
    f=np.fft.fftshift(np.fft.fft2(image))
    return 20*np.log(np.abs(f)+10**-6)

def func1(nx,ny,fx,fy,multiplier):
    return np.sin(np.pi/180*(2*np.pi*fx*nx*multiplier*1/max(fx,fy)))

def func2(nx,ny,fx,fy,multiplier):
    return np.sin(np.pi/180*(2*np.pi*fx*nx*multiplier*1/max(fx,fy)) + np.pi/180*(2*np.pi*fy*ny*multiplier*1/max(fx,fy)))

def func3(nx,ny,fx,fy,multiplier):
    return 10+np.cos(np.pi/180*(2*np.pi*multiplier*(fx*nx+fy*ny)))
    
def image_construct(id,func,multiplier):
    fx=np.random.choice(fxy)
    fy=np.random.choice(fxy)
    image=[]
    for i in range(N):
        row=[]
        for j in range(N):
            row.append(Q(func(i,j,fx,fy,multiplier)))
        image.append(row)
    plt.figure(figsize=(15,15))
    plt.subplot(121)
    img=plt.imshow(image)
    plt.title("Image "+ str(id)+" with\nfx="+str(fx)+"\nfy="+str(fy)+"\nMultiplier="+str(multiplier))
    plt.subplot(122)
    img=plt.imshow(dft_computation(image))
    plt.title("Mangitude spectrum")
    plt.show()
```

```python
image_construct(1,func1,10)
```

```python
image_construct(1,func1,2)
```

```python
image_construct(2,func2,10)
```

```python
image_construct(2,func2,2)
```

```python
image_construct(3,func3,10)
```

```python
image_construct(3,func3,2)
```

Aliasing effect is visible due to undersampling. DFT is more intense for images with differing pixel colors while the DFT is less intense for images with similar pixel colors

## Question 2

```python
block_sizes=[2,5,10,50]
a,b=0,255
for bs in block_sizes:
    image=[]
    for i in range(100):
        row=[]
        if(i%bs==0):
            a,b=b,a
        for j in range(100):
            if j%bs==0:
                a,b=b,a
            row.append(a)
        image.append(row)
    plt.figure(figsize=(15,15))
    plt.subplot(121)
    plt.title("Checkerboard with block size= "+str(bs)+"*"+str(bs))
    plt.imshow(image)
    plt.subplot(122)
    plt.title("Magnitude Spectrum")
    plt.imshow(dft_computation(image))
    plt.show()
```

DFT is more intense when colors are less differing in the image
